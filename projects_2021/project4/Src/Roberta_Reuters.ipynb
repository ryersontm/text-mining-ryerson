```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("/content/drive/MyDrive/BERT_canadian_AI/reuters_train.csv")
test = pd.read_csv("/content/drive/MyDrive/BERT_canadian_AI/reuters_test.csv")
```

```python
from simpletransformers.classification import ClassificationModel

#Create a ClassificationModel
model = ClassificationModel('roberta', 'roberta-base', num_labels=90, use_cuda = False)
```

```python
model.train_model(train)
```

```python
scores1, model_outputs, wrong_predictions = model.eval_model(test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(test['labels'].values,np.argmax(model_outputs,axis=1)))
```
